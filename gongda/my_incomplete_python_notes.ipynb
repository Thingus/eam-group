```python
# string

str = "London Marylebone NOx"
print(str)

# slice the string
print(str[0],str[-1],sep = ",")
print(str[0:9])
print(str[:-3])
print(str[0:9],str[20:23])
print(type(str),len(str),end=' ') # print results in one line 

# check if element "L" is in the string
print("L" in str)                

# string.split(separator, maxsplit) 

# example 
DateTime = "2016-01-01 12:59:59"

Date  = DateTime.split(" ")[0]
Time  = DateTime.split(" ")[1]
Year  = Date.split("-")[0]
Month = Date.split("-")[1]
Day   = Date.split("-")[2]
Hour  = Time.split(":")[0]
Min   = Time.split(":")[1]
Sec   = Time.split(":")[2]

print(Year,Month,Day,Hour,Min,Sec, sep = "   ")

# reverse the above example 
date = [Year,Month,Day]
sep = '-'
date = sep.join(date)

time = [Hour,Min,Sec]
sep = ':'
time = sep.join(time)

datetime = date + " " + time
print(datetime)

# string replace
str_1 = "London Marylebone NOx"
str_2 = str_1.replace("NOx","NO2")
print(str_2)
```

```python
# list 

# example 0
surface_NO2 = [] # empty list 

# example 1
a = [1,2,3,4]
a[2] = "x"       # assign a[2] = x
b = a + [5,6,7]
c = [x*2 for x in b]

for x in [a,b,c]:
    print(x)
    
# example 2
a = ['Monday','Tuesday','Wednesday','Friday','Saturday']
a.insert(3,'Thursday')   # add "Thursday" as a[3]
del a[0:2]               # delete a[0], a[1]
a.remove('Saturday')    # remove a certain element
a.append('Sunday')        # add to the end
print(a)

b = ['Monday','Tuesday','Wednesday','Friday','Saturday']
print(b[2:4])
print(b[-2:])

# example 3
a = [1,2,5,6,3,2,2,0]
b = sorted(a)       # sorted() creates a new list while keeping the original one
a.sort()            # .sort() sorts the list and changes the original one
a.reverse()

print(a,b)
b.count(2)         
```

```python
# dictionary 

# structure: {key: value} "key" must be Integer or String - "value" can be anything

# example 1
gas = {'NO2': 'nitrogen dioxide',
       'SO2': 'sulfur dioxide'}

print(gas)
print(gas.keys())
print(gas.values())

# example 2
short_name = ['NO2','SO2','CO','O3','PM2.5']
long_name  = ['nitrogen dioxide', 'sulfur dioxide', 'carbon monoxide', 'ozone', 'particulate matter 2.5']

pollutants = {}

for i in range(len(short_name)):
    dict1 = "{}".format(short_name[i])
    dict2 = "{}".format(long_name[i])
    pollutants[dict1]=dict2
    
print(pollutants)

pollutants['HCHO'] = 'formaldehyde' # add one more pollutant
del pollutants['NO2']               # remove NO2
print(pollutants['SO2'])            # look up SO2 
print('SO2' in pollutants)          # check if SO2 is in dictionary
print(pollutants)                   
```

```python
#####################################################################################################
# Flow control
# for loops - list comprehension
# if else
# while - continue - break

# example 1
a = [1,2,3,4,5,6,7,8]
b = []
c = []

for i in range(len(a)):
    if a[i] > 3 and a[i] < 6:
        b.append(a[i])
    else:
        b.append('NA')

for i in range(len(a)):
    if a[i] > 3 or a[i] < 6:
        c.append(a[i])
    else:
        c.append('NA')

print(b)
print(c)

# if all([condition_1,..., condition_n):
# if any([condition_1,..., condition_n):

# example 2
a = [-2,-1,0,1,2]
b = []

for i in range(len(a)):
    if a[i] < -1:
        b.append('too low')
    elif -1 <= a[i] < 0:
        b.append('low')
    elif a[i] == 0:
        b.append('perfect')
    elif 1 <= a[i] < 2:
        b.append('high')
    else:
        b.append('too high')


c = ['too low' if i < -1 else 'low' if -1 <= i < 0 
                         else 'perfect' if i == 0 
                         else 'high' if 1 <= i <2
                         else 'too high'for i in a]

d = [i*100 if i < -1 else i*200 if -1 <= i < 0 
                     else i*300 if i == 0 
                     else i*200 if 1 <= i <2
                     else i*100 for i in a]

for x in [a,b,c,d]:
    print(x)

# example 3 
i = 0
while i < 5:
    print(i, end = "  ")
    i = i+1

for i in range(15):
    if i%5 == 0:
        continue
    print(i, end = "/ ")

for i in range(15):
    print(i, end = "  ")
    if i >= 5:
        break
```
